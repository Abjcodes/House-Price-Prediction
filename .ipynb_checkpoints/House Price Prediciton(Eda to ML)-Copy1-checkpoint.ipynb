```python
None_type = ('BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'FireplaceQu','GarageType', 'GarageFinish', 'GarageQual', 'GarageCond','PoolQC','Fence', 'MiscFeature', 'Alley')

for col in None_type:
    df_train[col] = df_train[col].fillna('None')
    df_test[col] = df_test[col].fillna('None')
```

```python
for c in df_train:
    if df_train[c].dtype == 'object':
        df_train[c] = df_train[c].fillna(df_train[c].mode()[0])   
    else:
        df_train[c] = df_train[c].fillna(df_train[c].median())
```

```python
for c in df_test:
    if df_test[c].dtype == 'object':
        df_test[c] = df_test[c].fillna(df_test[c].mode()[0])   
    else:
        df_test[c] = df_test[c].fillna(df_test[c].median())
```

# Label Encoding

```python
def label_encoding(df_train, df_test):
    count = 0
    for c in df_train:
        if df_train[c].dtype == 'object':
            if len(list(df_train[c].unique())) <= 2:
                le = LabelEncoder()
                le.fit(list(df_train[c].unique())+list(df_test[c].unique()))
                df_train[c]= le.transform(df_train[c].astype(str))
                df_test[c]= le.transform(df_test[c].astype(str))
                count += 1
    print("Number of Label Encoded Columns : %d" %count)
```

# Feature Selection and Model Building

```python
corr = corrmat.sort_values('SalePrice')
cols = corr['SalePrice'][corr['SalePrice'].values > 0.2].index.values
```

```python
cols = np.delete(cols, len(cols) - 1)
train_sample = train[cols]
test_sample = test[cols]
```

```python
scaler = StandardScaler()
scaler.fit(train_sample)

train_sample = scaler.transform(train_sample)
test_sample = scaler.transform(test_sample)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
train_label = df_train['SalePrice']
X_train, X_test, y_train, y_test = train_test_split(train_sample, train_label, train_size = 0.8)
```

```python
model = LinearRegression()
model.fit(train_sample, train_label)
```

```python
prediction = model.predict(X_test)
```

```python
from sklearn import metrics
np.sqrt(metrics.mean_squared_error(y_test, prediction))
```

# Generating prediction on test dataset

```python
test_pred = model.predict(test_sample)
testID = df_test['Id']
submit = pd.DataFrame()
submit['ID'] = testID
submit['SalePrice'] = test_pred
submit.head(30)
```
